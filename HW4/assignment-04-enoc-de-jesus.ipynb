9. Git merge merges branch_1 onto the branch your in, and both branches are still in tact, along with the merge history being perserved as the commits are separate. 
git rebase doesn't keep the histories in tact. In git rebase, you are in branch_1, and you tell where you want to rebase, for example
get checkout branch_1
git rebase main
Git rebase then removes the commits you made on branch_1, resets it, and then the commits are reapplied on top of the main branch. This in turn rewrites the commit history, and no merge commit is shown.

So instead of merging the branch onto main while knowing when the merge commit happened, preserving branch history, you make a linear commit history in main as if no merge happened.
